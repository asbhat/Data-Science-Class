```python
# pip install lxml
# pip install cssselect
from bs4 import BeautifulSoup
from urllib2 import urlopen
from time import sleep
```

```python
food_url = ("http://www.urbanspoon.com/n/3/231/NYC/Gramercy-Flatiron-restaurants")
```

```python
def make_soup(url):
    html = urlopen(url).read()
    return BeautifulSoup(html, "lxml")
```

```python
website = make_soup(food_url)
```

```python
print website
```

```python
names = website.find_all("a", class_="resto_name")
ratings = website.find_all("span", class_="rating-percent")
addresses = website.find_all("span", class_="address")
food_types = website.find_all("span", class_="cuisines")
print names
```

```python
# Empty list of restaurants
restaurants = []
```

```python
for (name,rating,address,food_type) in zip(names,ratings,addresses,food_types):
    restaurant = {}
    restaurant['name'] = name.text
    restaurant['rating'] = rating.text
    restaurant['address'] = address.text
    restaurant['food_type'] = food_type.text
    restaurants.append(restaurant)
print restaurants
```

```python
import re
str = "90% like it"
number = re.findall(r'\d+', str)
number = " ".join(number)
print number
```

```python
import re
for restaurant in restaurants:
    restaurant['rating'] = " ".join(re.findall(r'\d+', restaurant['rating']))
print restaurants
    
```

```python
# Print to csv
output_file = open('restaurants.csv', 'w')
output_file.write('Name,Rating Percent,Food Type,Address')
for restaurant in restaurants:
    print "{0},{1},{2},{3}\n".format(restaurant['name'].strip(), restaurant['rating'].strip(), restaurant['food_type'].strip(), restaurant['address'].strip())
    output_file.write("{0},{1},{2},{3}\n".format(restaurant['name'].strip(), restaurant['rating'].strip(), restaurant['food_type'].strip(), restaurant['address'].strip()))
```
