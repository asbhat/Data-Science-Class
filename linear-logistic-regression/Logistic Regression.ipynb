```python
import pandas as pd
data = pd.read_csv('/Users/aditya/Github/Data-Science-Class/sentiment-model/training.txt', sep='\t')
print data
```

```python
data['love'] = data.Phrase.map(lambda x: 1 if 'love' in x else 0)
data
```

```python
import statsmodels.formula.api as sm

model = sm.logit('Sentiment ~ love', data=data).fit()
print model.summary()
```

____
**Note:** use tilde between 'Sentiment' and 'love'  
5.3848 represents that the sentence is 5.3x more likely to be positive if 'love' is included

```python
data['like'] = data.Phrase.map(lambda x: 1 if 'like' in x else 0)
data
```

```python
model = sm.logit('Sentiment ~ like', data=data).fit()
print model.summary()
```

```python
model = sm.logit('Sentiment ~ like + love + like*love', data=data).fit()
print model.summary()
```

```python
from patsy import dmatrices

y, X = dmatrices('Sentiment ~ like + love + like*love', data=data, return_type='matrix')
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X, y)

from sklearn.cross_validation import cross_val_score
print cross_val_score(model, X, y)
```
