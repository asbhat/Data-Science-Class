```python
datadf['lowdebt'] = datadf.DebtRatio.map(lambda x:1 if x < 0.01 else 0)
datadf['middebt'] = datadf.DebtRatio.map(lambda x:1 if 0.01 <= x and x < 100 else 0)
datadf['highdebt'] = datadf.DebtRatio.map(lambda x:1 if x >= 100 else 0)
```

```python
datadf.head()
```

```python
datadf['lowdebtvalue'] = datadf.DebtRatio * datadf.lowdebt
datadf['middebtvalue'] = datadf.DebtRatio * datadf.middebt
datadf['highdebtvalue'] = datadf.DebtRatio * datadf.highdebt
```

```python
datadf.head()
```

```python
imputer = Imputer(missing_values=0, strategy='mean', axis=0)
data_matrix = imputer.fit_transform(datadf[['lowdebtvalue', 'middebtvalue','highdebtvalue']])
```

```python
data_matrix
```

```python
datadf = np.column_stack([[datadf, data_matrix]])
```
