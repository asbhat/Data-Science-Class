```python
no14RawData = rawData[rawData['Operator'] != 'Partner14']
```

```python
no14RawData.head()
```

```python
set(no14RawData['Operator'])
```

```python
no14RawData['HH Deliv - Pur'] = no14RawData['Deliv HH Imps'] - no14RawData['Pur Ratecard HH Imps']
```

```python
matchedData = no14RawData[(no14RawData['Purchased Units'] == 1) & (no14RawData['Delivered Units'] == 1)]
```

```python
matchedData.head()
```

```python
ggplot(matchedData, aes(x='Pur Ratecard HH Imps', y='Deliv HH Imps')) + geom_point() + stat_smooth(color='blue') + xlab('Purchased Imps') + ylab('Delivered Imps') 
```

```python
ggplot(matchedData, aes(x='Pur Ratecard HH Imps', y='Deliv HH Imps')) + geom_point() + stat_smooth(color='blue') + xlab('Purchased Imps') + ylab('Delivered Imps') + facet_wrap('Quarter', scales='fixed')
```

```python
ggplot(matchedData, aes(x='Pur Ratecard HH Imps', y='Deliv HH Imps')) + geom_point() + stat_smooth(color='blue') + xlab('P Imps') + ylab('') + facet_wrap('Clean Demo 1 Name')
```
